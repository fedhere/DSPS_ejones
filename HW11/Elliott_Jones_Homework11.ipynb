# **Plotting each time series**

```
plt.plot(first[0], first[1], ',')
plt.title("The First TESS Light Curve")
plt.xlabel("Barycentric Julian Date (Days)")
plt.ylabel("Differential Magnitude (mmag)")
plt.show()

plt.plot(second[0], second[1], ',')
plt.title("The Second TESS Light Curve")
plt.xlabel("Barycentric Julian Date (Days)")
plt.ylabel("Differential Magnitude (mmag)")
plt.show()

plt.plot(third[0], third[1], ',')
plt.title("The Third TESS Light Curve")
plt.xlabel("Barycentric Julian Date (Days)")
plt.ylabel("Differential Magnitude (mmag)")
plt.show()
```

```
fourier_1 = LombScargle(first[0], first[1], normalization='psd').autopower()
fourier_2 = LombScargle(second[0], second[1], normalization='psd').autopower()
fourier_3 = LombScargle(third[0], third[1], normalization='psd').autopower()
```

# **Computing Lomb-Scargle periodograms for each time series**

```
plt.plot(fourier_1[0], fourier_1[1])
plt.xscale('log')
# plt.yscale('log')
plt.ylabel("Magnitude power", fontsize="large")
plt.xlabel("Period", fontsize="large")
plt.title("First time series periodgram")
plt.show()

plt.plot(fourier_2[0], fourier_2[1])
plt.xscale('log')
# plt.yscale('log')
plt.ylabel("Magnitude power", fontsize="large")
plt.xlabel("Period", fontsize="large")
plt.title("Second time series periodgram")
plt.show()

plt.plot(fourier_3[0], fourier_3[1])
plt.xscale('log')
# plt.yscale('log')
plt.ylabel("Magnitude power", fontsize="large")
plt.xlabel("Period", fontsize="large")
plt.title("Third time series periodgram")
plt.show()
```

# **Binning** 

```
import scipy as sp
import scipy.stats as stats
means_1, edges_1, binnum_1 = stats.binned_statistic(
    first_folded.phase, first_folded.dmag, bins=20)
means_2, edges_2, binnum_2 = stats.binned_statistic(
    second_folded.phase, second_folded.dmag, bins=20)
means_3, edges_3, binnum_3 = stats.binned_statistic(
    third_folded.phase, third_folded.dmag, bins=20)
```

```
plt.plot(first_folded.phase, first_folded.dmag, ',', zorder=0)
plt.hlines(means_1, edges_1[:-1], edges_1[1:], 'k', lw=4, zorder=1)
plt.xlabel("Period", fontsize = "large")
plt.ylabel("Magnitude", fontsize = "large")
plt.title("1st time series phase-folded with bins")
plt.show()
plt.plot(second_folded.phase, second_folded.dmag, ',', zorder=0)
plt.hlines(means_2, edges_2[:-1], edges_2[1:], 'k', lw=4, zorder=1)
plt.xlabel("Period", fontsize = "large")
plt.ylabel("Magnitude", fontsize = "large")
plt.title("Second time series phase-folded with bins")
plt.show()
plt.plot(third_folded.phase, third_folded.dmag, ',', zorder=0)
plt.hlines(means_3, edges_3[:-1], edges_3[1:], 'k', lw=4, zorder=1)
plt.xlabel("Period", fontsize = "large")
plt.ylabel("Magnitude", fontsize = "large")
plt.title("Third time series phase-folded with bins")
plt.show()
```
